```python
# Supprimer les colonnes Unnamed
df_mars = df_mars.drop(df_mars.columns[26:30], axis=1)

# Supprimer les lignes correspondant à la catégorie 'mini bus' dans la colonne 'carrosserie'
df_mars = df_mars.drop(df_mars[df_mars['Carrosserie'] == 'MINIBUS'].index)

# Supprimer les colonne champ_v9, date_maj et hybride
df_mars = df_mars.drop(['hybride','champ_v9', 'date_maj'], axis = 1)

# Réinitialiser les index après la suppression
df_mars = df_mars.reset_index(drop=True)

# Afficher le DataFrame après la suppression des lignes correspondant à la catégorie 'mini bus'
print(df_mars)
```

```python
# Remplacement virgule par point et conversion des colonnes en variable float
num_columns = ['puiss_max','conso_urb', 'conso_exurb', 'conso_mixte', 'co_typ_1', 'hc', 'nox', 
               'hcnox', 'ptcl']
for i in num_columns:
    df_mars[i] = df_mars[i].str.replace(',', '.')     #Remplacement des virgules par des points
    df_mars[i] = df_mars[i].astype('float')       #Conversion de la colonne en décimales

    
cor = df_mars.iloc[:,11:21].corr() # sélection des variables quantitatives 
# et calcul des corrélations entre ces variables
fig, ax = plt.subplots(figsize = (12,12))
sns.heatmap(cor, annot = True, ax = ax, cmap = "coolwarm")
plt.show()
```

```python
"""Il ya Une corrélation de 0.97 entre l'émission de Co2 et 3 variable : conso_mixte, conso_exurb et conso_urb
cela indique une corrélation linéaire positive parfaite (les variables augmentent ensemble)"""
```

```python
# Sélection des colonnes de type float et int
Num_cols = df_mars.select_dtypes(include=['float64', 'int64']).columns

# Remplacement des valeurs float et int manquantes par la médiane de chaque colonne
df_mars[Num_cols] = df_mars[Num_cols].fillna(df_mars[Num_cols].median())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import f1_score
#from imblearn.over_sampling import RandomOverSampler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer

# Pour éviter d'avoir les messages warning
import warnings
warnings.filterwarnings('ignore')
```

```python
# Supprimer les variables trop corellées avec la variable cible
df_mars = df_mars.drop(['conso_urb', 'conso_exurb', 'conso_mixte'], axis = 1)

# Réinitialiser les index après la suppression
df_mars = df_mars.reset_index(drop=True)
```

```python
df_mars.info()
```

```python
# Séparation du jeu de données sur la variable cible "co2"
X = df_mars.drop('co2', axis=1)
y = df_mars['co2']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state = 42)
```

```python
df_mars.shape
```

```python
# Encodage des variables catégorielles

from sklearn.preprocessing import LabelEncoder

var_cat = ['lib_mrq', 'lib_mod_doss', 'lib_mod', 'dscom', 'cnit', 'tvv', 'cod_cbr', 'typ_boite_nb_rapp', 'Carrosserie', 'gamme']

label_encoder = LabelEncoder()
for column in var_cat:
    X_train[column] = label_encoder.fit_transform(X_train[column])
    X_test[column] = label_encoder.fit_transform(X_test[column])

```

```python
X_train.head()
```

```python
# Normalisation des données
scaler = StandardScaler()
var_num = ['puiss_max','co_typ_1', 'hc', 'nox', 
               'hcnox', 'ptcl']
X_train.loc[:,var_num] = scaler.fit_transform(X_train[var_num])
X_test.loc[:,var_num] = scaler.transform(X_test[var_num])
```

```python
# Instancier un modèle LinearRegression de régression linéaire
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
# Afficher l'intercept ainsi que les coefficients de chaque variable estimée par le modèle
coeffs = list(regressor.coef_)
coeffs.insert(0, regressor.intercept_)

X2 = list(X.columns)
X2.insert(0, 'intercept')

pd.DataFrame({'valeur estimée': coeffs}, index=X2)
```

```python
# Afficher le score du modèle sur le jeu de données de d'entraînement et de test
print('Coefficient de détermination du modèle sur train:', regressor.score(X_train, y_train))
print('Coefficient de détermination du modèle sur test:', regressor.score(X_test, y_test))
```

```python
fig = plt.figure(figsize = (10,10))
pred_test = regressor.predict(X_test)
plt.scatter(pred_test, y_test, c='green')

plt.plot((y_test.min(), y_test.max()), (y_test.min(), y_test.max()), color = 'red')
plt.xlabel("prediction")
plt.ylabel("vrai valeur")
plt.title('Régression Linéaire pour les emissions de CO2')

plt.show();
```

```python
# Instancier le modèle de régression par arbre de décision
from sklearn.tree import DecisionTreeRegressor 

regressor = DecisionTreeRegressor(random_state=42) 
  
regressor.fit(X_train, y_train)

print(regressor.score(X_train, y_train))
print(regressor.score(X_test ,y_test))
```

```python
"""Un score de 1.0 sur les données d'entraînement indique une performance parfaite du modèle sur ces données, 
ce qui peut indiquer un surapprentissage (overfitting) du modèle """
```

```python
from sklearn.linear_model import LinearRegression

# Instanciation du modèle de régression linéaire
modele_regression_lineaire = LinearRegression()

# Entraînement du modèle sur les données d'entraînement
modele_regression_lineaire.fit(X_train, y_train)

# Évaluation du modèle
score_train = modele_regression_lineaire.score(X_train, y_train)
score_test = modele_regression_lineaire.score(X_test, y_test)

print("Score sur les données d'entraînement:", score_train)
print("Score sur les données de test:", score_test)
```

```python
# Calcule de l'importance de chacune des variables

import matplotlib.pyplot as plt

feat_importances = pd.DataFrame(regressor.feature_importances_, index=X.columns, columns=["Importance"])
feat_importances.sort_values(by='Importance', ascending=False, inplace=True)
feat_importances.plot(kind='bar', figsize=(8,6))
plt.show()
```

```python
from sklearn.ensemble import RandomForestRegressor 

regressor = RandomForestRegressor(random_state=42) 
  
regressor.fit(X_train, y_train)

print(regressor.score(X_train, y_train))
print(regressor.score(X_test, y_test))
```

```python
"""Dans l'ensemble, les résultats suggèrent que le modèle RandomForestRegressor est performant pour la tâche de régression, 
mais il est toujours important de surveiller la possibilité de surajustement et de rechercher des moyens de l'atténuer."""
```

```python
import sklearn.metrics
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

### DecisionTree

regressor_decision_tree = DecisionTreeRegressor(random_state=42) 
  
regressor_decision_tree.fit(X_train[var_num], y_train)


y_pred_decision_tree = regressor_decision_tree.predict(X_test[var_num])
y_pred_train_decision_tree = regressor_decision_tree.predict(X_train[var_num])

# Calcul des métriques

# jeu d'entraînement 
mae_decision_tree_train = mean_absolute_error(y_train,y_pred_train_decision_tree)
mse_decision_tree_train = mean_squared_error(y_train,y_pred_train_decision_tree,squared=True)
rmse_decision_tree_train = mean_squared_error(y_train,y_pred_train_decision_tree,squared=False)

# jeu de test 
mae_decision_tree_test = mean_absolute_error(y_test,y_pred_decision_tree)
mse_decision_tree_test = mean_squared_error(y_test,y_pred_decision_tree,squared=True)
rmse_decision_tree_test = mean_squared_error(y_test,y_pred_decision_tree,squared=False)


### RandomForest

regressor_random_forest = RandomForestRegressor(random_state=42) 
  
regressor_random_forest.fit(X_train[var_num], y_train)

# Calcul des métriques 
y_pred_random_forest = regressor_random_forest.predict(X_test[var_num])
y_pred_random_forest_train = regressor_random_forest.predict(X_train[var_num])


# jeu d'entraînement 
mae_random_forest_train = mean_absolute_error(y_train,y_pred_random_forest_train)
mse_random_forest_train = mean_squared_error(y_train,y_pred_random_forest_train,squared=True)
rmse_random_forest_train = mean_squared_error(y_train,y_pred_random_forest_train,squared=False)

# jeu de test 
mae_random_forest_test = mean_absolute_error(y_test,y_pred_random_forest)
mse_random_forest_test = mean_squared_error(y_test,y_pred_random_forest,squared=True)
rmse_random_forest_test = mean_squared_error(y_test,y_pred_random_forest,squared=False)


# Creation d'un dataframe pour comparer les metriques des deux algorithmes 
data = {'MAE train': [mae_decision_tree_train, mae_random_forest_train],
        'MAE test': [mae_decision_tree_test, mae_random_forest_test],
        'MSE train': [mse_decision_tree_train,mse_random_forest_train],
        'MSE test': [mse_decision_tree_test,mse_random_forest_test],
        'RMSE train': [rmse_decision_tree_train, rmse_random_forest_train],
        'RMSE test': [rmse_decision_tree_test, rmse_random_forest_test]}
  
# Creer DataFrame
df = pd.DataFrame(data, index = ['Decision Tree', 'Random Forest '])

df.head()

```

```python
"""la validation croisée avec Ridge et Lasso pour choisir les meilleurs hyperparamètres 
et évaluer la performance des modèles :"""

from sklearn.model_selection import cross_val_score
from sklearn.linear_model import Ridge, Lasso

# Définition des listes d'alphas à tester
alphas = [0.1, 0.5, 1, 5, 10]

# Initialisation des listes pour stocker les scores moyens de validation croisée pour Ridge et Lasso
ridge_cv_scores = []
lasso_cv_scores = []

# Validation croisée pour Ridge
for alpha in alphas:
    ridge_model_cv = Ridge(alpha=alpha)
    scores = cross_val_score(ridge_model_cv, X_train, y_train, cv=5)  # Utilisation de 5 plis de validation croisée
    ridge_cv_scores.append(scores.mean())

# Validation croisée pour Lasso
for alpha in alphas:
    lasso_model_cv = Lasso(alpha=alpha)
    scores = cross_val_score(lasso_model_cv, X_train, y_train, cv=5)
    lasso_cv_scores.append(scores.mean())

# Affichage des scores moyens de validation croisée pour chaque alpha
print("Scores moyens de validation croisée pour Ridge:", ridge_cv_scores)
print("Scores moyens de validation croisée pour Lasso:", lasso_cv_scores)

# Choix des meilleurs alphas pour Ridge et Lasso
best_alpha_ridge = alphas[np.argmax(ridge_cv_scores)]
best_alpha_lasso = alphas[np.argmax(lasso_cv_scores)]

print("\nMeilleur alpha pour Ridge:", best_alpha_ridge)
print("Meilleur alpha pour Lasso:", best_alpha_lasso)
```

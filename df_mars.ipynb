```python
import pandas as pd
import numpy as np
```

```python
df_mars = pd.read_csv("mars-2014-complete_01.csv", sep=";", header=0)

df_mars.head(10)
```

```python
df_mars.info()
```

```python
# VÃ©rifier les valeurs nulles dans le Data
valeurs_nulles = df_mars.isnull().sum()
print(valeurs_nulles)
```

```python
# Afficher un rÃ©sumÃ© statistique du Data
print(df_mars.describe())
```

```python
# Supprimer les lignes correspondant Ã  la catÃ©gorie 'mini bus' dans la colonne 'carrosserie'
df_mars = df_mars.drop(df_mars[df_mars['Carrosserie'] == 'MINIBUS'].index)

# RÃ©initialiser les index aprÃ¨s la suppression
df_mars = df_mars.reset_index(drop=True)

# Afficher le DataFrame aprÃ¨s la suppression des lignes correspondant Ã  la catÃ©gorie 'mini bus'
print(df_mars)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

sns.set() # pour modifier le thÃ¨me

sns.relplot(x = 'cod_cbr', y = 'co2', kind = 'line', data = df_mars, ci = None, height=8, color='red')

plt.xlabel('Carburant')
plt.ylabel('Emission de Co2\n')
plt.title("Relation entre le type de carburant et les Ã©missions de CO2\n")

plt.show()
```

```python
"""AprÃ¨s avoir examinÃ© les donnÃ©es et reprÃ©sentÃ© graphiquement la relation entre le type de carburant 
et les Ã©missions de CO2, il est clair que cette relation est significative 
pour les carburants tels que l'essence, le gazole et Super-Ã©thanol."""
```

```python
# Grouper les donnÃ©es par type de carburant et calculer la moyenne des Ã©missions de CO2 pour chaque groupe
mean_co2_by_fuel = df_mars.groupby('cod_cbr')['co2'].mean()

# Tracer le diagramme Ã  barres
plt.figure(figsize=(10, 6))  
mean_co2_by_fuel.plot(kind='bar', color='skyblue')  

# Ajouter des Ã©tiquettes et un titre
plt.xlabel('Type de carburant')
plt.ylabel('Ã‰mission moyenne de CO2')
plt.title('Ã‰mission moyenne de CO2 par type de carburant');
```

```python
"""Le diagramme Ã  barres illustrant l'Ã©mission moyenne de CO2 par type de carburant met en Ã©vidence des diffÃ©rences 
marquÃ©es entre les diffÃ©rents types. 
Pour l'essence et le gazole, par exemple, il est observÃ© que l'Ã©mission moyenne de CO2 est distincte, 
avec des valeurs plus Ã©levÃ©es pour l'essence par rapport au gazole."""
```

```python
# Tracer un scatter plot
sns.scatterplot(x='Carrosserie', y='co2', data=df_mars)

# Ajouter des Ã©tiquettes et un titre
plt.xlabel('Type de Carrosserie')
plt.ylabel('Ã‰missions de CO2')
plt.title('Relation entre les Ã©missions de CO2 et le type de Carrosserie\n')

# Faire pivoter les Ã©tiquettes de l'axe x pour une meilleure lisibilitÃ© (optionnel)
plt.xticks(rotation=45)

# Afficher le graphique
plt.show()
```

```python
"""les donnÃ©es analysÃ©es soulignent l'effet significatif du type de carrosserie sur les Ã©missions de CO2, 
ce qui met en Ã©vidence l'importance de considÃ©rer ce facteur dans le choix 
des consommateurs en matiÃ¨re de vÃ©hicules."""
```

```python
df_mars['puiss_max'] = df_mars['puiss_max'].replace(',', '.').astype(float)

sns.lmplot(x='puiss_max', y='co2', lowess=True, data=df_mars)

plt.title("Relation entre la puissance maximale et les Ã©missions de CO2\n")


plt.show()
```

```python
"""Il est important de noter que la relation entre la puissance maximale et les Ã©missions de CO2 peut varier 
d'un modÃ¨le de vÃ©hicule Ã  l'autre et d'un fabricant Ã  l'autre, en fonction des spÃ©cifications du moteur, 
des technologies de contrÃ´le des Ã©missions et des stratÃ©gies de conception spÃ©cifiques."""
```

```python
df_mars = df_mars.astype(str)
df_mars.iloc[:, 11:21] = df_mars.iloc[:, 11:21].apply(lambda x: x.str.replace(',', '.')).astype(float)

cor = df_mars.iloc[:,11:21].corr() # sÃ©lection des variables quantitatives 
# et calcul des corrÃ©lations entre ces variables
fig, ax = plt.subplots(figsize = (12,12))
sns.heatmap(cor, annot = True, ax = ax, cmap = "coolwarm")
plt.show()
```

```python
"""Il ya Une corrÃ©lation de 0.97 entre l'Ã©mission de Co2 et 3 variable : conso_mixte, conso_exurb et conso_urb
cela indique une corrÃ©lation linÃ©aire positive parfaite (les variables augmentent ensemble)"""
```

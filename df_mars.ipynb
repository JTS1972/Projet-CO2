```python
import pandas as pd
import numpy as np
```

```python
df_mars = pd.read_csv("mars-2014-complete_01.csv", sep=";", header=0)

df_mars.head(10)
```

```python
df_mars.info()
```

```python
# Vérifier les valeurs nulles dans le Data
valeurs_nulles = df_mars.isnull().sum()
print(valeurs_nulles)
```

```python
# Afficher un résumé statistique du Data
print(df_mars.describe())
```

```python
# Supprimer les lignes correspondant à la catégorie 'mini bus' dans la colonne 'carrosserie'
df_mars = df_mars.drop(df_mars[df_mars['Carrosserie'] == 'MINIBUS'].index)

# Réinitialiser les index après la suppression
df_mars = df_mars.reset_index(drop=True)

# Afficher le DataFrame après la suppression des lignes correspondant à la catégorie 'mini bus'
print(df_mars)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

sns.set() # pour modifier le thème

sns.relplot(x = 'cod_cbr', y = 'co2', kind = 'line', data = df_mars, ci = None, height=8, color='red')

plt.xlabel('Carburant')
plt.ylabel('Emission de Co2\n')
plt.title("Relation entre le type de carburant et les émissions de CO2\n")

plt.show()
```

```python
"""Après avoir examiné les données et représenté graphiquement la relation entre le type de carburant 
et les émissions de CO2, il est clair que cette relation est significative 
pour les carburants tels que l'essence, le gazole et Super-éthanol."""
```

```python
# Grouper les données par type de carburant et calculer la moyenne des émissions de CO2 pour chaque groupe
mean_co2_by_fuel = df_mars.groupby('cod_cbr')['co2'].mean()

# Tracer le diagramme à barres
plt.figure(figsize=(10, 6))  
mean_co2_by_fuel.plot(kind='bar', color='skyblue')  

# Ajouter des étiquettes et un titre
plt.xlabel('Type de carburant')
plt.ylabel('Émission moyenne de CO2')
plt.title('Émission moyenne de CO2 par type de carburant');
```

```python
"""Le diagramme à barres illustrant l'émission moyenne de CO2 par type de carburant met en évidence des différences 
marquées entre les différents types. 
Pour l'essence et le gazole, par exemple, il est observé que l'émission moyenne de CO2 est distincte, 
avec des valeurs plus élevées pour l'essence par rapport au gazole."""
```

```python
# Tracer un scatter plot
sns.scatterplot(x='Carrosserie', y='co2', data=df_mars)

# Ajouter des étiquettes et un titre
plt.xlabel('Type de Carrosserie')
plt.ylabel('Émissions de CO2')
plt.title('Relation entre les émissions de CO2 et le type de Carrosserie\n')

# Faire pivoter les étiquettes de l'axe x pour une meilleure lisibilité (optionnel)
plt.xticks(rotation=45)

# Afficher le graphique
plt.show()
```

```python
"""les données analysées soulignent l'effet significatif du type de carrosserie sur les émissions de CO2, 
ce qui met en évidence l'importance de considérer ce facteur dans le choix 
des consommateurs en matière de véhicules."""
```

```python
df_mars['puiss_max'] = df_mars['puiss_max'].replace(',', '.').astype(float)

sns.lmplot(x='puiss_max', y='co2', lowess=True, data=df_mars)

plt.title("Relation entre la puissance maximale et les émissions de CO2\n")


plt.show()
```

```python
"""Il est important de noter que la relation entre la puissance maximale et les émissions de CO2 peut varier 
d'un modèle de véhicule à l'autre et d'un fabricant à l'autre, en fonction des spécifications du moteur, 
des technologies de contrôle des émissions et des stratégies de conception spécifiques."""
```

```python
df_mars = df_mars.astype(str)
df_mars.iloc[:, 11:21] = df_mars.iloc[:, 11:21].apply(lambda x: x.str.replace(',', '.')).astype(float)

cor = df_mars.iloc[:,11:21].corr() # sélection des variables quantitatives 
# et calcul des corrélations entre ces variables
fig, ax = plt.subplots(figsize = (12,12))
sns.heatmap(cor, annot = True, ax = ax, cmap = "coolwarm")
plt.show()
```

```python
"""Il ya Une corrélation de 0.97 entre l'émission de Co2 et 3 variable : conso_mixte, conso_exurb et conso_urb
cela indique une corrélation linéaire positive parfaite (les variables augmentent ensemble)"""
```

```python
x = df_mars['co2']
y = df_mars['typ_boite_nb_rapp']

plt.figure(figsize=(8, 6))  # Définir la taille de la figure

plt.scatter(x, y, color='blue', alpha=0.5)  # Tracer le nuage de points

plt.xlabel('Co2')
plt.ylabel('type de boîte à vitesse\n')
plt.title('Diagramme de dispersion entre émissions de Co2 et le type de boîte à vitesse\n')

plt.show()
```

```python
#D'après la matrice de corrélation, les variables les plus corrélées avec Co2 sont conso_mixte, conso_exurb et conso_urb.
sns.pairplot(data = df_mars[['co2','conso_mixte','conso_exurb', 'conso_urb']])
plt.show()
```

```python
sns.lmplot(x = "co2", y = "conso_mixte", lowess = True, data = df_mars)
sns.lmplot(x = "co2", y = "conso_exurb", lowess = True, data = df_mars)
sns.lmplot(x = "co2", y = "conso_urb", lowess = True, data = df_mars)

plt.show()
```

```python
# Histogramme de distribution du CO2
plt.figure(figsize=(8, 6))
sns.histplot(data=df_mars, x='co2', bins=30, kde=True)
plt.title('Distribution des émissions de CO2')
plt.xlabel('CO2 (g/km)')
plt.ylabel('Fréquence')
plt.show()

"""Histogramme de distribution du CO2 : Montrant la distribution des émissions de CO2 pour tous les véhicules. 
Cela permettra de visualiser la répartition des niveaux d'émissions et d'identifier les valeurs extrêmes."""
```

```python
# Diagramme à barres pour comparer les émissions de CO2 par type de véhicule
plt.figure(figsize=(10, 6))
sns.barplot(data=df_mars, x='Carrosserie', y='co2', errorbar=None)
plt.title('Émissions de CO2 par type de véhicule')
plt.xlabel('Type de véhicule')
plt.ylabel('CO2 (g/km)')
plt.xticks(rotation=45)
plt.show()

"""Diagramme à barres pour comparer les émissions de CO2 par type de véhicule : 
Utilisez un diagramme à barres pour comparer les émissions de CO2 moyennes ou médianes entre les différents types 
de véhicules (par exemple, essence, diesel, hybride). Cela peut aider à identifier les types de véhicules les plus polluants."""
```
